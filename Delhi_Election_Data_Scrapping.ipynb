Here is the below code we have scrap the data from the Election commission site(http://results.eci.gov.in/DELHITRENDS2020/ConstituencywiseU0570.htm?ac=70). Since We hve 70 constituency that's why we have to run the loop 70 times for getting the data from each constituency and add it in the csv file

```python
from bs4 import BeautifulSoup
import urllib.request as urllib2
import csv

f = open('delhi_assembly_2020.csv','w+')
#We are adding header in the file delhi_assembly_2020 file

f.write("\t".join(["sno","CandidateName","Party_Name","EVM Votes","Postal Votes","Total Votes","%Votes","constuency"]))
f.write("\n")

# Running loop for getting the data from 70 constiuency
state_assembly_cnt = range(1,71)
for assembly_cnt in state_assembly_cnt:
    assembly_cnt = str(assembly_cnt)
    #We are making the url for scrapping the data from the election commission data.
    url = "http://results.eci.gov.in/DELHITRENDS2020/ConstituencywiseU05{}.htm?ac={}".format(assembly_cnt,assembly_cnt)
    #print(url)
    html = urllib2.urlopen(url)
    soup = BeautifulSoup(html.read(),"lxml")
    
    #print(soup)
    #We are extracting data for getting the constituency name here.
    
    s = soup.find('td',  {"align":"center", "colspan":"9"})    
    constuency = str(s.text.strip()).replace('NCT OF Delhi-','')
    #print(constuency)
    
    # we are traversing the each row and the extracting the each columns and extract the data into in a list and then moving
    #each record to the file and saving the all data
    
    for tr in soup.find_all("tr", {"style":"font-size:12px;"}):
        td = [td for td in tr.stripped_strings]
        td.append(constuency)
        #print("\t".join(td))
        f.write("\t".join(td))
        f.write("\n")
f.close()
```

Above script has extracted the constiuency data. Now we need a file to get the loksabha constuency wise data. Below we are scrapping the data from wikipedia and creating other file for legislative assembly constituency wise loksabha constituency mapping.

```python
from bs4 import BeautifulSoup
import urllib.request as urllib2
import csv
f = open('delhi_legislative_assembly_Loksabha.csv','w+')
url = "https://en.wikipedia.org/wiki/List_of_constituencies_of_the_Delhi_Legislative_Assembly"
html = urllib2.urlopen(url)
soup = BeautifulSoup(html.read(),"lxml")

f.write("\t".join(["sno","assembly_constituency","Loksabha_constituency" ]))
f.write("\n")
#print(soup)
table = soup.find('table',{"class":"wikitable sortable"})
#print(table)
for tr in table.find_all('tr'):
    td = [td for td in tr.stripped_strings]
    lst=[td[0],td[1],td[-1]]
    if "Constituency" in td[0]:
        continue
    print("\t".join(lst))
    f.write("\t".join(lst))
    f.write("\n")
f.close()
```

In below we are creating list of member of parliament based on the constituency

```python
from bs4 import BeautifulSoup
import urllib.request as urllib2
import csv
f = open('delhi_mplist.csv','w+')
url = "http://delhiassembly.nic.in/delhimps/delhimps.htm"
html = urllib2.urlopen(url)
soup = BeautifulSoup(html.read(),"lxml")
table = soup.find('table',{"align":"center", "border":"2"})
#print(table)
f.write("\t".join(["sno","Constituency","Member","Party"]))
f.write("\n")
for tr in table.find_all('tr',{"style":"font-family: Arial, Helvetica, sans-serif; font-size: small; font-weight: normal; font-style: normal; font-variant: normal; color: #000000"}):
    td = [td for td in tr.stripped_strings]
    if "National Capital Territory" in td[1]:
        continue
    f.write("\t".join(td))
    f.write("\n")
f.close()
```
